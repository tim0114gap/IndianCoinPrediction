```
from keras.models import load_model, model_from_json
import numpy as np
from PIL import Image
from keras.preprocessing.image import img_to_array, load_img
```

```
# load Pic
filename = "testPic/"
pic_name = filename + "rupee10.jpeg"
img = Image.open(pic_name)
img = img.resize((224, 224))
img = img.convert('RGB')
data = np.asarray(img)

# scaling
data = data / 255

# to 4dim
data = data[None, ...]
print(data.shape)
```

```
# save models and weights
model = model_from_json(open('DSProject_mizumashi2_model.json').read())
model.load_weights('DSProject_mizumashi2_model_weights.h5')
```

```
rupeeList = ['rupee_2', 'rupee_10', 'rupee_1', 'rupee_5']
features = model.predict(data)
print(rupeeList)
round(features[0][0], 3), round(features[0][1], 3), round(features[0][2], 3), round(features[0][3], 3)
```

```
feature = [round(features[0][0], 3), round(features[0][1], 3), round(features[0][2], 3), round(features[0][3], 3)]
maxNum = 0
num = 0
for i in range(len(feature)):
    if maxNum < feature[i]:
        maxNum = feature[i]
        num = i
print(rupeeList[num]," : ",maxNum)
```
